#Import the required libraries

```python
import sklearn
# import Logistic regression for sklearn
from sklearn.linear_model import LogisticRegression
# The dataset is pre-built
from sklearn import datasets
iris = datasets.load_iris()
```

```python
# Assign the features and target to separate variables
X = iris.data
y = iris.target
```

```python
# Visualising the dataset
#import library
import matplotlib.pyplot as plt

# Separate the features
features = iris.data.T
sepal_length = features[0]
sepal_width = features[1]
petal_length = features[2]
petal_width = features[3]

# setting labels for the plot figure
sepal_length_label = iris.feature_names[0]
sepal_width_label = iris.feature_names[1]
petal_length_label = iris.feature_names[2]
petal_width_label = iris.feature_names[3]

# Here we choose a scatter graph to visualise sepal width and length
plt.scatter(sepal_width, sepal_length, c=iris.target)
plt.xlabel(sepal_width_label)
plt.ylabel(sepal_length_label)

#print the graph
plt.show()
```

#Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state=0)
```

#Build a Model with K-Nearest Neighbors (KNN)

```python
from sklearn import neighbors
# empty model
classifier = neighbors.KNeighborsClassifier()
#fit to actual training data
classifier.fit(X_train, y_train)
```
